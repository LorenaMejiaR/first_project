```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/clean/final_df.csv')

df.head()
df
```

```python
df['area'].unique()
```

```python
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
    'Denmark', 'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 
    'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 
    'Lithuania', 'Luxembourg', 'Malta', 'Norway', 'Poland', 'Portugal', 
    'Romania', 'Serbia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 
    'Ukraine', 'United Kingdom'
]
```

### Create new dataframe only with the countries in the above list
- Drop 'Unnamed: 0' column
- Reset index

```python
eu_df = df[df['area'].isin(european_countries)]
eu_df = eu_df.drop('Unnamed: 0', axis=1) # this one was a repeated column with the index numbers
eu_df = eu_df.reset_index(drop=True)

eu_df
```

```python
print(eu_df.dtypes) #check data types
```

### Calculate correlation between R&D and HDI
- `.corr()` computes the linear relationship between variables
1. Group data by 'area'
2. calculate the correlation using .corr()

```python
# Filter the dataset based on non-zero standard deviations for 'r_d' and 'hdi'
df_clean = eu_df.groupby('area').filter(lambda x: x['r_d'].std() > 0 and x['hdi'].std() > 0)

# Calculate correlation for each 'area'
country_correlations = df_clean.groupby('area')[['r_d', 'hdi']].apply(lambda x: x['r_d'].corrcoef(x['hdi']))

# reset index for easier merge
country_correlations = country_correlations.reset_index(name='rd_hdi_correlation')

# merge correlation values back
final_df = pd.merge(df_clean, country_correlations, on='area', how='left')

# Drop duplicates if you only want one row per 'area' and 'year'
# final_df = final_df.drop_duplicates(subset=['area', 'year'])

# round correlations to 2 decimal
final_df['rd_hdi_correlation'] = final_df['rd_hdi_correlation'].round(3)

# keep relevant columns
final_df = final_df[['area', 'year', 'r_d', 'hdi', 'rd_hdi_correlation']]
final_df

# that took a minute, but that's the result (decimals rounded to 3)
# I'm happy to share the code because I found very weird that all give 1 or -1
# yes 
```

### Line Plot with Facets by Country (Area)

```python
sns.set(style="whitegrid")

# create a line plot
g = sns.FacetGrid(final_df, col="area", col_wrap=4, height=4)
g.map(sns.lineplot, "year", "r_d", color="blue", label="R&D Spending")
g.map(sns.lineplot, "year", "hdi", color="orange", label="HDI")

g.add_legend()
plt.show()
```

### Scatter Plot: R&D vs HDI for Each Year

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=eu_df, x='r_d', y='hdi', hue='area', palette="viridis", style='year', s=150)
plt.title('R&D vs HDI (Color-coded by Year)')
plt.xlabel('R&D Investment')
plt.ylabel('Human Development Index')
plt.legend(title='Year')
plt.show()
```

### Heatmap of `rd_hdi_correlation Across Countries and Years

```python
# Pivot data to get correlation by country and year
correlation_pivot = eu_df.pivot_table(values='rd_hdi_correlation', index='area', columns='year')

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_pivot, annot=True, cmap='Spectral', center=0, fmt='.2f')
plt.title('R&D vs HDI Correlation by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

### Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivoted DataFrame where we have countries as rows and the years as columns for r_d and hdi
df_pivot_r_d = eu_df.pivot_table(index='area', columns='year', values='r_d')
df_pivot_hdi = eu_df.pivot_table(index='area', columns='year', values='hdi')

# Combine the two DataFrames (r_d and hdi) into one
df_combined = pd.concat([df_pivot_r_d, df_pivot_hdi], axis=1, keys=['r_d', 'hdi'])

# Calculate the correlation matrix for the combined DataFrame
corr = df_combined.corr()

# Set the size of the figure
plt.figure(figsize=(14, 10))

# Create the heatmap using a nicer color palette (viridis, cividis, etc.)
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f', linewidths=1, linecolor='black', cbar=True)

# Title and axis labels
plt.title('Heatmap: Correlation between R&D, HDI, and Year by Country', fontsize=16)
plt.xlabel('Year and Variable', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Make sure the country names are visible and not rotated
plt.yticks(rotation=0)

# Display the plot
plt.show()

### the funny thing is that when I generated the maps, SOMEHOW the numbers are different???
# we might need to double check the .corr() function then, it might not be the right way of doing it.
# also some countries were returning NaN 
# .corr() theoretically returns the linear correlation between values, but 
# I was trying to display the country names here, that's what I was doing when the call started. I can work further in it.
#It should? We'll see.
```
