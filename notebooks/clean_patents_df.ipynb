```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
patents_df = pd.read_csv('../data/raw/SYB66_264_202310_Patents.csv', encoding='ISO-8859-1')
patents_df.columns = patents_df.columns.str.lower().str.replace(' ', '_')
patents_df
```

```python
patents_df.columns
```

```python
patents_df = patents_df.drop(index=0).reset_index(drop=True)
patents_df = patents_df.rename(columns={
    't29': 'country_code',
    'patents': 'area',
    'unnamed:_2': 'year',
    'unnamed:_3': 'series',
    'unnamed:_4': 'value',
    'unnamed:_5': 'footnotes',
    'unnamed:_6': 'source'
})
patents_df
```

```python
patents_df = patents_df.drop(columns= ['footnotes', 'source'])
patents_df
```

```python
patents_df.to_csv('clean_patents_df.csv', index=False)
```
